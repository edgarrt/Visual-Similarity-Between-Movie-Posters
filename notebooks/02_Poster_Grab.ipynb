# Movie Posters Fetch

```python
import pandas as pd
import requests
import os
from dotenv import load_dotenv
import numpy as np
import tqdm
```

## Data Related

```python
root = os.path.dirname(os.path.abspath(os.getcwd()))
```

```python
image_url = "https://image.tmdb.org/t/p/original/"
```

```python
data = pd.read_csv(os.path.join(root, "data", "raw_data.csv"), lineterminator='\n', index_col=0)
```

```python
data
```

## Download Movie Posters

```python
def download_poster(poster_path):
    to_download = f"{image_url}{poster_path}"
    img_data = requests.get(to_download).content
    with open(os.path.join(root, "images", "raw", poster_path), 'wb') as handler:
        handler.write(img_data) 
```

```python
# roughly ~30 minutes for full download
for index, row in tqdm.tqdm(data.iterrows(), total=data.shape[0]):
    download_poster(row["poster_path"])
```

```python
# in case initial download timed out
need = []
for index, row in tqdm.tqdm(data.iterrows(), total=data.shape[0]):
    file = f"{root}/images/raw/{row['poster_path']}"
    if not os.path.exists(file):
        need.append(row["poster_path"])
```

```python
need
```

```python
# if need to download missing images
for image in tqdm.tqdm(need):
    download_poster(image)
```
